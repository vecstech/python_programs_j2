# file handling in python
1. file is a collection of information and stored in memory
2. files are 2 types `1. text files` and `2. binary files`
3. text files are `.txt, .py, .html, .ipynb etc..`
4. binary files are `.exe, .apk, .mp4 etc..`

# files operations
1. create file - w mode
2. write file - w mode
3. read file - r mode
4. append file - a mode

```python
#create a file using python script
create_file=open("profile.txt","w")
print("file created successfully in home directory")
create_file.close()
```

```python
#write data to files using python script
write_file=open("profile.txt","w")
data="welcome to texas international academy \nlearning python program"
write_file.write(data)
print("data write successfully and verify the profile.txt file")
write_file.close()
```
