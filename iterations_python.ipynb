```python
# writea program to print 1 to 100 numbers using for loop and while loop?
for num in range(1,101):
    print(num,end=",")
```

```python
sn=1
en=100
while (sn<=en):
    print(sn,end=",")
    sn=sn+1
```

```python
# write a program to print even numbers in between 500 to 700 using for loop & while loop?
for n in range(500,701):
    if (n%2==0):
        print(n,end=",")
```

```python
sn=500
en=700
while (sn<=en):
    if (sn%2==0):
        print(sn,end=",")
    sn=sn+1
```

```python
# write a program to print sum of all even numbers and odd numbers using for loop and while loop?
se=0
for i in range(1,101):
    if (i%2==0):
        se=se+i
print("sum of all even numbers is",se)
```

```python
sn=1
en=100
se=0
while (sn<=en):
    if (sn%2==0):
        se=se+sn
    sn=sn+1
print("sum of all even numbers is",se)
```

```python
#write a program to print sum of all odd numbers in between 1 to 100 using for loop and while loop?
so=0
for i in range(1,101):
    if (i%2==1):
        so=so+i
print("sum of all odd numbers is",so)
```

```python
sn=1
en=100
so=0
while (sn<=en):
    if (sn%2==1):
        so=so+sn
    sn=sn+1
print("sum of all odd numbers is",so)
```
