# List in data structures
1. list is collection of elements enclosed by []
2. list a mutable - editable
3. it's allowed duplicate elements
4. each element seperated by ,

```python
#create list
names=["shiva","vikram","vikas","alok"]
ages=[25,26,27,25.5,26.7,25,26]
print(names)
print(ages)
print(type(names))
print(type(ages))
```

```python
#concatenation of list
a=[10,20,30,40,50]
b=[50,40,60,70,80]
c=a+b
print(c)
```

```python
#repeating list
numbers=[10,20,30,40,50]
print(numbers*5)
```

```python
#accessing elements from list - index 
names=["shiva","vikram","vikas","divya"]
print(names)
print(names[2])
print(names[-3])
```

```python
# methods or function in list
print(dir(list))
```

```python
# description
print(help(list))
```

```python
# adding elements to the list - append(), extend(), insert()
numbers=[10,20,30,40,50]
print(numbers)
numbers.append(80)
print(numbers)
```

```python
numbers=[10,20,30,40,50]
print(numbers)
numbers.extend([60,70,90])
print(numbers)
```

```python
numbers=[10,20,30,40,50]
print(numbers)
numbers.insert(1,100)
print(numbers)
```

```python
#pop(),remove()
numbers=[10,20,30,40,50]
print(numbers)
numbers.pop(1)
print(numbers)
```

```python
numbers=[10,20,30,40,50]
print(numbers)
numbers.remove(20)
print(numbers)
```

```python
#sort() - ascending , descending
numbers=[1,6,2,7,3,4,5]
print(numbers)
numbers.sort(reverse=True) # Fasle Asc - True Desc
print(numbers)
```

# tuples in data structres
1. tuple is collection of elemenst enclosed by ()
2. tuple is immutabe - not editable
3. each element seperated ,
4. duplicates allowed


```python
#create tuple
t1=(10,20,30,40,50,20,30)
print(t1)
print(type(t1))
```

```python
#concatenation, repeating, accessing elements
```

```python
#function or methods in tuples
print(dir(tuple))
```
