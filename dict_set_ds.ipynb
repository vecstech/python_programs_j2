# dict in data structures
1. dict is collection of items enclosed by {} - here item is a key:value pair
2. in dict keys are unique

```python
# create dict
info={"name":"shiva","add":"hyd","salary":15000,"age":25}
print(info)
print(type(info))
```

```python
# accesing values from dict using keys
info={"name":"shiva","add":"hyd","salary":15000,"age":25}
print(info)
print(info["salary"])
```

```python
#methods or function in dict
print(dir(dict))
```

```python
#keys(),values(),items()
info={"name":"shiva","add":"hyd","salary":15000,"age":25}
print(info)
print(info.keys())
print(info.values())
print(info.items())
```

```python
#update()
info={"name":"shiva","add":"hyd","salary":15000,"age":25}
print(info)
info["add"]="hyderabad"
print(info)
info["domine"]="data engineer"
print(info)
```

```python
#update()
info={"name":"shiva","add":"hyd","salary":15000,"age":25}
print(info)
info.update({"add":"hyderabad"})
print(info)
info.update({"domine":"data engineer"})
print(info)
```

```python
#pop()
info={"name":"shiva","add":"hyd","salary":15000,"age":25}
print(info)
info.pop("add")
print(info)
```

# sets in data structures
1. set is collection of unorder unique elements enclosed by {}

```python
#create set
a={70,50,15,30,20,10,10,30,15}
print(a)
print(type(a))
```

```python
#methods in set
print(dir(set))
```

```python
#add() - add elements to set
a={10,20,30,45,60}
print(a)
a.add(100)
print(a)
```

```python
#difference
a={1,2,3,5,4}
b={3,4,5,6,7,8}
c=b.difference(a)
print(c)
```

```python
#union()
a={1,2,3,5,4}
b={3,4,5,6,7,8}
c=a.union(b)
print(c)
```

```python
#intersection
a={1,2,3,5,4}
b={3,4,5,6,7,8}
c=a.intersection(b)
print(c)
```

```python
# discard()
a={10,20,30,40,50}
print(a)
a.discard(30)
print(a)
```
