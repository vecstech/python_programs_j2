# functions in Python
1. function is set of statements executed by calling the function name
2. functions are 2 types `1.pre defined functons` and `2.user defined functions`
3. pre define functions are `print(), type(), list(), tuple(), dict(), int(), input() ...etc`
4. user defined functions created by user
``` python user defined functions

    statement1..
```

````python
# inbuilt functions
print("print the outputs")
print(sum([1,2,3,4,5,6,7]))
print(pow(2,6))
print(sorted([6,4,5,1,3,2]))
print(dir(tuple))
print(help(list))
````

````python
# user define functions
# write a program to print addition of two numbers using function
def add_numbers():
    a=int(input("enter first number:"))
    b=int(input("enter second number:"))
    result=a+b
    print(result)
    return result
add_numbers()
````

````python
# write a program to print addition of two strings using function
def add_strings():
    a=str(input("enter first string:"))
    b=str(input("enter second string:"))
    result=a+b
    print(result)
    return result
add_strings()
````

````python
#write a program to print profile using functions with arguments
def profile(name,location,salary):
    print("my name is",name)
    print("my address is",location)
    print("my salary is",salary)
profile(name="vikram",location="hyderabad",salary=50000)
profile(name="shiva",location="banglore",salary=100000)
````
