1. python have 4 types of conditional statements
2. `if` statement
3. `elif` statement
4. `else` statement
5. `nested` statement

```python - if
if (condtion): #if condition True inside statements are running
    statement1
    statements...
```

```python - if,else
if (condition): #if condition True inside statements are running
    statement1
    statements2...n
else: #if conditon Fails its went to else statement and print inside statements

```

```python - if,elif,else
if (condtion):
    statements
elif (condition):
    statements
elif (condition):
    statements
else:
    statements
```

````python
# write a program to find the given number is even number?
number=int(input("enter number here:"))
if (number%2 == 0):
    print(number,"is even number")
````

````python
# write a program to find the given number is even number or odd number?
num = int(input("please enter number here:"))
if (num%2 == 1):
    print(num,"is odd number")
else:
    print(num,"is even number")
````

````python
# write a program to find the given number is positive or negitive or zero?
n = int(input("please enter number here:"))
if (n>0):
    print(n,"is positive number")
elif (n<0):
    print(n,"is negitive number")
else:
    print(n,"is zero")
````

````python
#write a program to find the largest number from three number ?
v1=int(input("enter first number here:"))
v2=int(input("enter second number here:"))
v3=int(input("enter third number here:"))
if (v1>=v2 and v1>=v3):
    print(v1,"is largest number")
elif (v2>=v1 and v2>=v3):
    print(v2,"is the largest number")
else:
    print(v3,"is the largest number")
````

````python
"""
write a program to find the grades based on student marks
1. if marks in between 80 to 100 - A Grade
2. if marks in betwen 60 to 79 - B Grade
3. if marks in between 40 to 59 - C Grade
4. if marks in between 25 to 39 - D Grade
5. if marks in between 0 to 24 - Fail
6. if other than above marks output is - Please enter correct marks 
"""
marks = int(input("enter student marks here:"))
if (marks>=80 and marks<=100):
    print(marks,"A Grade")
elif (marks>=60 and marks<=79):
    print(marks,"B Grade")
elif (marks>=40 and marks<=59):
    print(marks,"C Grade")
elif (marks>=25 and marks<=39):
    print(marks,"D Grade")
elif (marks>=0 and marks<=24):
    print(marks,"is Fail")
else:
    print("please enter correct marks")
````
