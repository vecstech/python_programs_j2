# Strings in Data Structures
1. strings is a collection of information or charecters enclosed by " " or ' '

```python
# create a string
name="python program"
info='welcome to tia'
print(name)
print(info)
print(type(name))
print(type(info))
```

```python
#multi line string
data="""i am learning python program
python with data analysis
python with machine learning
python with artificial intelligence
"""
print(data)
print(type(data))
```

```python
#concatenation of string - adding two or more strings
a="python"
b="program"
print(a+b)
print(a+" "+b)
```

```python
#repeate string
data="machine Learning"
print(data)
print(data*10)
```

### accesing charecters from string - Index Based
#### index are two types
1. forward index - read charecters from left to right and count is (0,1,2,3,,,n-1)
2. Backward or negitive index - read chrecters from right to left and count is (-1,-2,-3,,,,,-n)

```python
name="Machine Learning"
print(name)
print(name[0])
print(name[-1])
```

```python
#slicing concept -read group of charecters - string[start_index:end_index-1]
name="machine learning"
print(name)
print(name[0:7])
print(name[8:16])
print(name[8:])
print(name[-8:])
```

```python
name="machine learning"
print(name)
print(name[::-1])
```
